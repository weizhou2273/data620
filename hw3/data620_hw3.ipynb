```python
# Import modules
import networkx as nx
import networkx.algorithms.bipartite as bipartite
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pandas as pd
import numpy as np
from scipy import stats
from pylab import rcParams
import math
%matplotlib inline
```

```python
G = nx.davis_southern_women_graph()
```

```python
G.nodes()
```

```python
# Visualize the 2-mode networks
women, events = nx.bipartite.sets(G)
plt.rcParams["figure.figsize"] = (16,8)
plt.axis('off')
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, nodelist=women, node_color='red', alpha = 0.4)
nx.draw_networkx_nodes(G, pos, nodelist=events, node_color='blue', alpha = 0.4, node_size = 1000)
nx.draw_networkx_edges(G, pos, alpha = 0.4)
nx.draw_networkx_labels(G, pos);
```

### Diameter
A diameter of 4 is the shortest path between the furthest away nodes in the graph.

```python
nx.diameter(G)
```

### Radius
The radius is the minimum eccentricity in a graph - nodes with eccentricities equal to the radius constitute the "center" of the graph. 
In this case Radius is just 1 less node than than the diameter, indicating the small range of this network.

```python
nx.radius(G)
```

### Degree

### Relationship among women

```python
# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
pd.DataFrame(np.array([[W.degree(w),w] for w in women]),\
             columns=['Friend number','Women']).\
                sort_values(['Friend number'],\
                ascending=False)
```

The more friends a women has, the more centricity the women will be in the network.

```python
# Above data shows the number of friends each woman has.  The following shows the detialed relationship amoung this group.
women = G.graph['top']
W = bipartite.weighted_projected_graph(G, women,ratio=False)
weights = [edata['weight'] for f,t,edata in W.edges(data=True)]
plt.rcParams["figure.figsize"] = (16,8)
nx.draw_networkx(W, k=0.1,iterations=20, width=weights, edge_color=weights)
plt.axis('off')
plt.show()
```

### Relationship among events

```python
eventss = G.graph['bottom']
E = bipartite.weighted_projected_graph(G, events,ratio=False)
weights = [edata['weight'] for f,t,edata in E.edges(data=True)]
plt.rcParams["figure.figsize"] = (16,8)
nx.draw_networkx(E, width=weights, edge_color=weights)
plt.axis('off')
plt.show()
```
